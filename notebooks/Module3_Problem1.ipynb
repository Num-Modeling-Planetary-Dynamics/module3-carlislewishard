```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import glob
```

```python
files = glob.glob("../data/*.csv")
dfs = []
id_num = []
for f in files:
    id_num.append(int(f.rsplit("id")[1].rsplit("-")[0].lstrip("0")))
    dfs.append(pd.read_csv(f))
for i in range(len(dfs)):
    dfs[i]['id'] = id_num[i]
df = pd.concat(dfs, sort=False)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

ax[0].scatter(df.xh, df.yh, c=df.id, cmap='rainbow', s=0.5)
ax[0].set_xlabel('XH (AU)', fontsize=12)
ax[0].set_ylabel('YH (AU)', fontsize=12)
ax[0].set_title('XY Position Over Time', fontsize=12)

ax[1].scatter(df.xh, df.zh, c=df.id, cmap='rainbow', s=0.5)
ax[1].set_xlabel('XH (AU)', fontsize=12)
ax[1].set_ylabel('ZH (AU)', fontsize=12)
ax[1].set_title('XZ Position Over Time', fontsize=12)

plt.tight_layout()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
filename_mer = f'../data/id000002-XV-LF.csv'
df_mer = pd.read_csv(filename_mer)
filename_ven = f'../data/id000003-XV-LF.csv'
df_ven = pd.read_csv(filename_ven)
filename_ear = f'../data/id000004-XV-LF.csv'
df_ear = pd.read_csv(filename_ear)
filename_mar = f'../data/id000005-XV-LF.csv'
df_mar = pd.read_csv(filename_mar)
filename_jup = f'../data/id000006-XV-LF.csv'
df_jup = pd.read_csv(filename_jup)
filename_sat = f'../data/id000007-XV-LF.csv'
df_sat = pd.read_csv(filename_sat)
filename_ura = f'../data/id000008-XV-LF.csv'
df_ura = pd.read_csv(filename_ura)
filename_nep = f'../data/id000009-XV-LF.csv'
df_nep = pd.read_csv(filename_nep)
filename_plu = f'../data/id000010-XV-LF.csv'
df_plu = pd.read_csv(filename_plu)
```

```python
def calc_energy(df):
    de = []
    for i in range(len(df)):
        r = np.sqrt(df['px'][i]**2 + df['py'][i]**2 + df['pz'][i]**2)
        v2 = df['vx'][i]**2 + df['vy'][i]**2 + df['vz'][i]**2
        e = (0.5 * v2) - (1 / r)
        if (i == 0):
            e0 = e
        de.append(np.abs(e0 - e) / e0)
    df['de'] = de
    
def calc_angmom(df):
    dh = []
    for i in range(len(df)):      
        h = np.sqrt(((df['py'][i] * df['vz'][i]) - (df['pz'][i] * df['vy'][i]))**2 + ((df['pz'][i] * df['vx'][i]) - (df['px'][i] * df['vz'][i]))**2 + ((df['px'][i] * df['vy'][i]) - (df['py'][i] * df['vx'][i]))**2)
        if (i == 0):
            h0 = h
        dh.append(np.abs(h0 - h) / h0)
    df['dh'] = dh

def plot(df):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
    
    ax[0].plot(df[' t'], df['de'])
    ax[0].set_xlabel('Time (years)', fontsize=12)
    ax[0].set_ylabel('Change in Energy', fontsize=12)
    ax[0].set_title('$\Delta \epsilon$', fontsize=12)
    ax[0].tick_params(axis='both',direction='in')
    
    ax[1].plot(df[' t'], df['dh'])
    ax[1].set_xlabel('Time (years)', fontsize=12)
    ax[1].set_ylabel('Change in Angular Momentum', fontsize=12)
    ax[1].set_title('$\Delta h$', fontsize=12)
    ax[1].tick_params(axis='both',direction='in')
    
    plt.tight_layout()
```

```python
calc_energy(df_mer)
calc_energy(df_ven)
calc_energy(df_ear)
calc_energy(df_mar)
calc_energy(df_jup)
calc_energy(df_sat)
calc_energy(df_ura)
calc_energy(df_nep)
calc_energy(df_plu)
```

```python
calc_angmom(df_mer)
calc_angmom(df_ven)
calc_angmom(df_ear)
calc_angmom(df_mar)
calc_angmom(df_jup)
calc_angmom(df_sat)
calc_angmom(df_ura)
calc_angmom(df_nep)
calc_angmom(df_plu)
```

```python
plot(df_mer)
plot(df_ven)
plot(df_ear)
plot(df_mar)
plot(df_jup)
plot(df_sat)
plot(df_ura)
plot(df_nep)
plot(df_plu)
```

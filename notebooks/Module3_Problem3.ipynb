```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
filename_old = f'../data/id000010-XV.csv'
df_old = pd.read_csv(filename_old)
filename_new = f'../data/id000010-XV-new.csv'
df_new = pd.read_csv(filename_new)
```

```python
rlist = []
vlist = []
tlist = df_old.values.tolist()
for i in range(len(df_old)):
    px = df_old['xh'][i] - df_new['px'][i]
    py = df_old['yh'][i] - df_new['py'][i]
    pz = df_old['zh'][i] - df_new['pz'][i]
    r = np.sqrt(px**2 + py**2 + pz**2)
    rlist.append(r)
    vx = df_old['vxh'][i] - df_new['vx'][i]
    vy = df_old['vyh'][i] - df_new['vy'][i]
    vz = df_old['vz'][i] - df_new['vz'][i]
    v = np.sqrt(vx**2 + vy**2 + vz**2)
    vlist.append(v)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
plt.plot(tlist, rlist)
plt.xlabel('Time (years)', fontsize=12)
plt.ylabel('r', fontsize=12)
plt.title('Difference in Position over Time', fontsize=12)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
plt.plot(tlist, vlist)
plt.xlabel('Time (years)', fontsize=12)
plt.ylabel('v', fontsize=12)
plt.title('Difference in Velocity over Time', fontsize=12)
plt.tight_layout()
```
